```
from bs4 import BeautifulSoup
import requests
```

```
import string
import random
import time
```

We scrape the website www.toutelaconjugaison.com to extract a list of all existing french verbs, using requests and BeautifulSoup4

```
alphabet = list(string.ascii_uppercase)
baseUrl = "https://www.toutelaconjugaison.com/lexique-lettre-LETTER-page-PAGENUMBER.html"
```

```
random.shuffle(alphabet)
```

```
def extractFrenchVerbs(letter = "A",pageNb=1,baseUrl="https://www.toutelaconjugaison.com/lexique-lettre-LETTER-page-PAGENUMBER.html"):
  '''
  Take french verbs from the website "toutelaconjugaison.com", given a letter and a page number.

  Parameters:
    letter (str): the letter the verbs are starting with.
    pageNb (int): the page number of the list.

  Returns (list): a list of french verbs starting with letter A on page pageNb.
  '''
  pageNb = str(pageNb)
  url = baseUrl.replace("LETTER",letter).replace("PAGENUMBER",pageNb)
  response = requests.get(url)
  html = BeautifulSoup(response.text,"html.parser")
  htmlEls = html.find_all(id="cadremilieu")[0].find_all("a")
  htmlEls = [el for el in htmlEls if not el.get("class")]
  return [verb.text.strip().lower() for verb in htmlEls]
```

```
frenchVerbs = []
for letter in alphabet:
  page = 1
  extractedVerbs = extractFrenchVerbs(letter=letter,pageNb=page,baseUrl=baseUrl)
  frenchVerbs.append(extractedVerbs)
  while extractedVerbs:
    print("letter : ",letter,", page : ",page,", ex : ",extractedVerbs[0])
    wait = random.randint(1,10)
    time.sleep(wait) #wait for some time in between requests.
    page += 1
    extractedVerbs = extractFrenchVerbs(letter=letter,pageNb=page,baseUrl=baseUrl)
    frenchVerbs.append(extractedVerbs)
```

```
frenchVerbsFlatten = [verb for verbs in frenchVerbs for verb in verbs]
```

```
with open(file="frenchVerbs.txt",mode="w") as f:
  for verb in frenchVerbsFlatten:
    f.write(verb + " \n")
```
